```python
import numpy as np
import matplotlib.pyplot as plt

def he(t, A, a):
    """
    Alpha : h_e(t) = A * a * t * e^(-at)
     t >= 0 H(t)
    """
    return A * a * t * np.exp(-a * t)

# 
t = np.linspace(0, 10, 500)

# 
#  1:  (Standard)
A1, a1 = 1.0, 1.0
y1 = he(t, A1, a1)

#  2: A  (A=2) -> 
A2, a2 = 2.0, 1.0
y2 = he(t, A2, a2)

#  3: a  (a=2) -> ()(a)
A3, a3 = 1.0, 2.0
y3 = he(t, A3, a3)

# 
plt.figure(figsize=(10, 6))

# 
plt.plot(t, y1, label=f'Standard (A={A1}, a={a1})', linewidth=2, color='blue')
plt.plot(t, y2, label=f'Larger A (A={A2}, a={a2})', linewidth=2, linestyle='--', color='green')
plt.plot(t, y3, label=f'Faster a (A={A3}, a={a3})', linewidth=2, linestyle='-.', color='red')

#  (Peak Points)
#  t = 1/a,  = A/e
peak_t1, peak_h1 = 1/a1, A1/np.e
peak_t2, peak_h2 = 1/a2, A2/np.e
peak_t3, peak_h3 = 1/a3, A3/np.e

plt.scatter([peak_t1, peak_t2, peak_t3], [peak_h1, peak_h2, peak_h3], color='black', zorder=5)
plt.text(peak_t1, peak_h1 + 0.02, f'Peak 1\n({peak_t1}, {peak_h1:.2f})', ha='center')
plt.text(peak_t2, peak_h2 + 0.02, f'Peak 2 (Higher)\n({peak_t2}, {peak_h2:.2f})', ha='center')
plt.text(peak_t3, peak_h3 + 0.02, f'Peak 3 (Faster)\n({peak_t3}, {peak_h3:.2f})', ha='center')

# 
plt.title(r'Alpha Function: $h_e(t) = A \cdot a \cdot t \cdot e^{-at}$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(0, 5)  #  x 
plt.ylim(0, 1.0)

plt.tight_layout()
plt.show()
```
